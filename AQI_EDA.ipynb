# Now let's calculate the AQI

#### Sub-Index = [(High AQI - Low AQI) / (High Breakpoint - Low Breakpoint)] × (Pollutant Concentration - Low Breakpoint) + Low AQI

```python
AQI=pd.read_html('https://github.com/auriga-arun/pythontraining/blob/main/Problem-1.md')[0]
```

```python
AQI
```

```python
df
```

```python
df.isnull().sum()
```

```python
breakpoints = {
    'PM2.5': {'0-50': (0, 30), '51-100': (31, 60), '101-200': (61, 90), '201-300': (91, 120), '301-400': (121, 250), '401-500': (251, float('inf'))},
    'PM10': {'0-50': (0, 50), '51-100': (51, 100), '101-200': (101, 250), '201-300': (251, 350), '301-400': (351, 430), '401-500': (431, float('inf'))},
    'NO2': {'0-50': (0, 40), '51-100': (41, 80), '101-200': (81, 180), '201-300': (181, 280), '301-400': (281, 400), '401-500': (401, float('inf'))},
    'SO2': {'0-50': (0, 40), '51-100': (41, 80), '101-200': (81, 380), '201-300': (381, 800), '301-400': (801, 1600), '401-500': (1601, float('inf'))},
    'CO': {'0-50': (0, 1), '51-100': (1.1, 2), '101-200': (2.1, 10), '201-300': (10.1, 17), '301-400': (17.1, 34), '401-500': (34.1, float('inf'))},
    'O3': {'0-50': (0, 50), '51-100': (51, 100), '101-200': (101, 168), '201-300': (169, 208), '301-400': (209, 748), '401-500': (749, float('inf'))}
}
```

```python
def calculate_subindex(pollutant,concentration):
    for aqi_range,(low_breakpoint,high_breakpoint) in breakpoints[pollutant].items():
        if (low_breakpoint<=concentration) and (concentration<=high_breakpoint):
            low_aqi,high_aqi=map(int,aqi_range.split('-'))
            return ((high_aqi - low_aqi) / (high_breakpoint - low_breakpoint)) * (concentration - low_breakpoint) + low_aqi
    print(concentration,pollutant)
```

```python
def funct(grp):
    pollutant=['PM2.5','PM10','NO2','CO','SO2','O3']
    subindex=[]
    for i in pollutant:
        subindex.append(calculate_subindex(i,round(grp[i])))
    return max(subindex)
        
```

```python
df['AQI']=df.apply(funct,axis=1)
```

```python
def f(g):
    for aqi_range in AQI['AQI Range']:
        lower,upper=map(int,aqi_range.split('-'))
        aqi=round(g['AQI'])
        if (lower<=aqi) and (aqi<=upper):
            return AQI[AQI['AQI Range']==aqi_range]['Category'].iloc[0]
    print('oho')
```

# How does air quality vary across seasons (Winter, Summer, Monsoon, Post-Monsoon)?

```python
df['Month'].unique()
```

```python
def funti(month):
    month=int(month)
    if (month==12) or ((month>=1 and month<=2)):
        return 'Winter'
    elif (month>=3 and month<=5):
        return 'Summer'
    elif (month>=6 and month<=9):
        return 'Monsoon'
    else:
        return 'Post-Monsoon'
```

```python
df['Seasons']=df['Month'].apply(funti)
```

```python
df
```

```python
season_data=df.groupby('Seasons')['AQI'].mean().reset_index()
```
